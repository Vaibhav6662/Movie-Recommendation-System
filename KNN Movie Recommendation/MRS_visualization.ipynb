```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
```

```python
ratings.rename(columns={"id": "movie_id"}, inplace=True)
```

```python
merged_df = pd.merge(movies, ratings, on="movie_id")
```

```python
top_rated = merged_df.sort_values(by="vote_average", ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="vote_average", y="title_x", data=top_rated, palette="viridis")
plt.title("Top 10 Highest Rated Movies", fontsize=16)
plt.xlabel("Average Rating")
plt.ylabel("Movie Title")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_df, x="vote_count", y="vote_average", hue="title_x", legend=False)
plt.title("Vote Count vs Average Rating", fontsize=16)
plt.xlabel("Vote Count")
plt.ylabel("Vote Average")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top_vote_count = merged_df.sort_values(by="vote_count", ascending=False).head(5)
plt.figure(figsize=(8, 8))
plt.pie(top_vote_count["vote_count"], labels=top_vote_count["title_x"], autopct="%1.1f%%", startangle=140)
plt.title("Top 5 Movies by Vote Count", fontsize=16)
plt.tight_layout()
plt.show()
```
